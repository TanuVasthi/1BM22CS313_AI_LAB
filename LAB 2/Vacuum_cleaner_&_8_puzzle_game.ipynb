```python
#vacuum cleaner
def vacuum_world():
    # Initializing goal_state
    goal_state = {'A': '0', 'B': '0'}  # 0 indicates Clean, 1 indicates Dirty
    cost = 0

    location_input = input("Enter Location of Vacuum (A/B): ").strip()
    status_input = input(f"Enter status of {location_input} (0 for Clean, 1 for Dirty): ").strip()
    status_input_complement = input("Enter status of other room (0 for Clean, 1 for Dirty): ").strip()

    print("Initial Location Condition: " + str(goal_state))

    if location_input == 'A':
        print("Vacuum is placed in Location A.")

        if status_input == '1':  # Location A is Dirty
            print("Location A is Dirty.")
            goal_state['A'] = '0'  # Clean A
            cost += 1  # Cost for cleaning A
            print("Cost for CLEANING A: " + str(cost))
            print("Location A has been Cleaned.")

        if status_input_complement == '1':  # If B is Dirty
            print("Location B is Dirty.")
            print("Moving right to Location B.")
            cost += 1  # Cost for moving right
            print("COST for moving RIGHT: " + str(cost))
            goal_state['B'] = '0'  # Clean B
            cost += 1  # Cost for cleaning B
            print("COST for SUCK: " + str(cost))
            print("Location B has been Cleaned.")

    elif location_input == 'B':
        print("Vacuum is placed in Location B.")

        if status_input == '1':  # Location B is Dirty
            print("Location B is Dirty.")
            goal_state['B'] = '0'  # Clean B
            cost += 1  # Cost for cleaning B
            print("COST for CLEANING B: " + str(cost))
            print("Location B has been Cleaned.")

        if status_input_complement == '1':  # If A is Dirty
            print("Location A is Dirty.")
            print("Moving left to Location A.")
            cost += 1  # Cost for moving left
            print("COST for moving LEFT: " + str(cost))
            goal_state['A'] = '0'  # Clean A
            cost += 1  # Cost for cleaning A
            print("COST for SUCK: " + str(cost))
            print("Location A has been Cleaned.")

    else:
        print("Invalid location entered.")

    # Done cleaning
    print("GOAL STATE: ")
    print(goal_state)
    print("Performance Measurement: " + str(cost))

# To run the function
vacuum_world()
```

```python
#8 puzzle game
from collections import deque


class GridProblem:
    def __init__(self, initial_state, goal_state, grid):
        # Initializes a grid problem instance with initial and goal states, and the grid layout
        self.initial_state = initial_state
        self.goal_state = goal_state
        self.grid = grid

    def is_goal(self, state):
        # Checks if the given state is the goal state
        return state == self.goal_state

    def is_valid_cell(self, row, col):
        # Checks if the given cell coordinates are within the grid boundaries and not blocked
        return (
            0 <= row < len(self.grid) and
            0 <= col < len(self.grid[0]) and
            self.grid[row][col] == 0
        )

    def expand(self, node):
        # Expands the given node by generating child nodes for valid adjacent cells
        row, col = node.state
        children = []
        for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            new_row, new_col = row + dr, col + dc
            if self.is_valid_cell(new_row, new_col):
                child_state = (new_row, new_col)
                child_node = Node(child_state, parent=node)
                children.append(child_node)
        return children


class Node:
    def __init__(self, state, parent=None, action=None):
        # Initializes a node with a state, parent node (optional), and action (optional)
        self.state = state
        self.parent = parent
        self.action = action


def breadth_first_search(problem):
    # Performs breadth-first search algorithm to find a solution for the given problem
    node = Node(problem.initial_state)
    if problem.is_goal(node.state):
        return node

    frontier = deque([node])
    reached = {problem.initial_state}

    while frontier:
        node = frontier.popleft()

        for child in problem.expand(node):
            state = child.state

            if problem.is_goal(state):
                return child
            if state not in reached:
                reached.add(state)
                frontier.append(child)
    return None


def reconstruct_path(node):
    # Reconstructs the path from the goal node back to the initial node
    path = []
    while node:
        path.append(node.state)
        node = node.parent
    return list(reversed(path))


def print_complete_path(path):
    # Prints the complete path from start to goal
    if path:
        for step, point in enumerate(path):
            print("Step {}: {}".format(step, point))
    else:
        print("No solution found")


# Define a 3x3 grid
# 1 : Denotes the obstacles
# 0 : Empty space or a non-obstacle cell in the grid
grid_3x3 = [
    [0, 0, 1],  # First row: two empty cells, one obstacle
    [0, 0, 0],  # Second row: all empty cells
    [1, 0, 0]   # Third row: one obstacle, two empty cells
]

# Define initial and goal states
initial_state_3x3 = (0, 0)  # Top-left corner
goal_state_3x3 = (2, 2)      # Bottom-right corner

# Define the problem instance
problem_3x3 = GridProblem(initial_state_3x3, goal_state_3x3, grid_3x3)

# Perform breadth-first search to find a solution
solution_node_3x3 = breadth_first_search(problem_3x3)

# Print solution if found
print('!! Reached the Goal!!' if solution_node_3x3 else None)
if solution_node_3x3:
    print("Solution found!")
    solution_path_3x3 = reconstruct_path(solution_node_3x3)
    print("Complete Path:")
    print_complete_path(solution_path_3x3)
else:
    print("No solution found")
```
